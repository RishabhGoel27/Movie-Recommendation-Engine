```python
def change(obj):
    lst=[]
    for i in ast.literal_eval(obj):
        lst.append(i['name'])
    return lst    
        
```

```python

movies['genres']=movies['genres'].apply(change)
```

```python
movies['keywords']=movies['keywords'].apply(change)
```

```python
movies.head()
```

```python
def changecast(obj):
    lst=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
             lst.append(i['name'])
             count+=1
        else:
            break
    return lst    
        
```

```python
def  choose(obj):
    lst=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            lst.append(i['name'])
    return lst    
        
```

```python
movies['crew']=movies['crew'].apply(choose)
```

```python
movies['cast']=movies['cast'].apply(changecast)
```

```python
def new_strip(obj):
    obj=obj.lstrip('tt0')
    return obj
```

```python
movies['imdb_id']=movies['imdb_id'].apply(new_strip)
```

```python
movies.rename(columns={"imdb_id":"imdbId"},inplace=True)
```

```python
movies['imdbId']=movies['imdbId'].astype(int)
```

```python
movies=movies.drop(['id','imdbId'],axis=1)
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
new=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tag']=new['tag'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tag']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x: x[1])
    for i in distances[1:11]:
        print(new.iloc[i[0]].title)
    
```

```python
recommend("Toy Story")
```
